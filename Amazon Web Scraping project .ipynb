```python
firefox_options = Options()

# Run in headless mode (no browser UI)
firefox_options.headless = True

# Create a new Firefox profile for performance settings
firefox_profile = webdriver.FirefoxProfile()

# Disable loading images for faster performance
firefox_profile.set_preference("permissions.default.image", 2)


# Disable extensions
firefox_profile.set_preference("extensions.enabledScopes", 0)

# Set the profile in the options
firefox_options.profile = firefox_profile

# Initialize WebDriver with the performance optimized options
driver = webdriver.Firefox(options=firefox_options)


def set_postal_code(zip_code):
    try:
        # Wait for the location button to be clickable
        location_button = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.ID, 'glow-ingress-block'))
        )
        location_button.click()

        # Wait for the postal code input to appear
        postal_code_input = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.ID, 'GLUXZipUpdateInput'))
        )

        # Clear existing postal code and enter the new one
        postal_code_input.clear()
        postal_code_input.send_keys(zip_code)

        # Click the apply button to set the postal code
        apply_button = driver.find_element(By.XPATH, '//*[@id="GLUXZipUpdate"]/span/input')
        apply_button.click()

        # Pause to allow for dynamic content loading
        time.sleep(5)

        print(f"Zip code set to {zip_code} and continued.")
        
    except TimeoutException:
        print(f"Error: Failed to set zip code {zip_code}. Element not found or took too long to load.")

    finally:
        print("Zip code set to 10001 and continued.")



```

```python
def Get_Pro_Price(product_details):
    """ Get the product price price details """

    try:
        product_price_nodis = driver.find_element(By.CLASS_NAME, 'a-price-whole')
        price_total = product_price_nodis.text.strip()
        product_details['Total Price (No Discount)'] = price_total
    except NoSuchElementException:
        product_details['Total Price (No Discount)'] = float('nan')


    try:
        product_price = driver.find_element(By.XPATH, '//*[@id="corePrice_desktop"]/div')
        price_text = product_price.text.strip()
    except NoSuchElementException:
        price_text = float('nan')

    product_details['List Price'] = float('nan')
    product_details['Current Price'] = float('nan')
    product_details['You Save'] = float('nan')

    # Assuming price_text is provided and could be None or a float
    if price_text is None or isinstance(price_text, float):
        price_text = ""

    # Extract List Price
    list_price_match = re.search(r'List Price:\s*\$(\d+\.\d{2})', price_text)
    if list_price_match:
        product_details['List Price'] = list_price_match.group(1)

    # Extract Current Price
    current_price_match = re.search(r'Price:\s*\$(\d+\.\d{2})', price_text)
    if current_price_match:
        product_details['Current Price'] = current_price_match.group(1)
    else:
        # If the "Price:" label isn't there, try matching any standalone price
        standalone_price_match = re.search(r'\$(\d+\.\d{2})', price_text)
        if standalone_price_match:
            product_details['Current Price'] = standalone_price_match.group(1)

    # Extract "You Save" amount
    you_save_match = re.search(r'You Save:\s*\$(\d+\.\d{2})', price_text)
    if you_save_match:
        product_details['You Save'] = you_save_match.group(1)
    

    try:
        stock_tag_0 = driver.find_element(By.XPATH, '//*[@id="availabilityInsideBuyBox_feature_div"]/div')
        stock_availability = stock_tag_0.text.strip()
        if stock_availability == 'In Stock':
            product_details['Stock Availability'] = 'Yes'
        elif stock_availability == 'Out of Stock':
            product_details['Stock Availability'] = 'No'
        else:
            product_details['Stock Availability'] = stock_availability
    except NoSuchElementException:
        product_details['Stock Availability'] = float('nan')
```

```python
def Get_Pro_Info(product_details):
    """Function to extract product information from the Amazon page."""


    try:
        #product_content_tag = driver.find_element(By.CSS_SELECTOR, '#productDescription p')
        product_content_tag = driver.find_element(By.CSS_SELECTOR, '#renewedProgramDescriptionAtf_feature_div')
        product_content = product_content_tag.text.strip()
    except NoSuchElementException:
        product_content = 'Product description not found'
    
    product_details['Product Description'] = product_content

    try:
        Title_content = driver.find_element(By.XPATH, '//*[@id="titleSection"]')
        product_details['Title Content'] = Title_content.text.strip()
    except NoSuchElementException:
        product_details['Title Content'] = float('nan')

    try:
        rate_tag = driver.find_element(By.XPATH, '//*[@id="acrPopover"]')
        rate = rate_tag.text.strip()
        product_details['Rating(out of 5)'] = rate
    except:
        product_details['Rating(out of 5)'] = float('nan')


    try:
        total_rate_tag = driver.find_element(By.XPATH, '//*[@id="averageCustomerReviews"]/span[3]')
        total_rate_text = total_rate_tag.text.strip()
        
        # Extract only the numbers using a regular expression
        total_rate_numbers = re.findall(r'\d+', total_rate_text)
        total_rate = ''.join(total_rate_numbers) if total_rate_numbers else float('nan')
        
        product_details['Total number of Rating'] = total_rate
    except:
        product_details['Total number of Rating'] = float('nan')

    try:
        about_item_tag = driver.find_element(By.XPATH, '//*[@id="feature-bullets"]')
        about_item = about_item_tag.text.strip().replace('\n', ' ')
        product_details['About Item'] = about_item
    except:
        product_details['About Item'] = float('nan')


    try:
        service_provider_tag = driver.find_element(By.XPATH, '//*[@id="variation_service_provider"]')
        product_details['Service Provider'] = service_provider_tag.text.strip()
    except NoSuchElementException:
        product_details['Service Provider'] = "Unlocked"

    if isinstance(product_details['Service Provider'], str):
        product_details['Service Provider'] = product_details['Service Provider'].replace('\n', ', ').replace('Service Provider: ', '').replace('.', '')



    prod_Info = driver.find_element(By.XPATH, '//*[@id="productDetails_feature_div"]').text
    fields = {
        'Form Factor': r'Form Factor\s+(.*)',
        'Manufacturer': r'Manufacturer\s+(.*)',
        'Whats in the box': r'Whats in the box\s+(.*)',
        'Battery Power Rating': r'Battery Power Rating\s+(.*)',
        'Other camera features': r'Other camera features\s+(.*)',
        'Human Interface Input': r'Human Interface Input\s+(.*)',
        'Other display features': r'Other display features\s+(.*)',
        'GPS': r'GPS\s+(.*)',
        'OS': r'OS\s+(.*)',
        'Batteries': r'Batteries\s+(.*)',
        'Item model number': r'Item model number\s+(.*)',
        'ASIN': r'ASIN\s+(.*)',
        'Item Weight': r'Item Weight\s+(.*)',
        'Product Dimensions': r'Product Dimensions\s+(.*)',
    }

    # Loop through each field and apply regex to extract data
    for field_name, regex in fields.items():
        match = re.search(regex, prod_Info)
        if match:
            product_details[field_name] = match.group(1).strip()
        else:
            if field_name == 'Human Interface Input':
                product_details[field_name] = "Touchscreen"
            elif field_name == 'GPS':
                product_details[field_name] = "TRUE"
            else:
                product_details[field_name] = float("nan")
            


    # Extract product title and add it to the dictionary
    try:
        product_title = driver.find_element(By.XPATH, '//*[@id="productTitle"]').text.strip()
        product_details['Title Content'] = product_title
    except NoSuchElementException:
        product_details['Title Content'] = "Title not found"



    try:

        # Find all review elements
        review_elements = driver.find_elements(By.CLASS_NAME, "a-row.a-spacing-small.review-data")
        reviews = []
        
        # Iterate over each review element
        for review_element in review_elements:
            review = review_element.find_element(By.TAG_NAME, "span").text
            review = review.replace("\n", "  ")
            review = review.replace("-", "")
            reviews.append(review)
        
        # Get the top comment
        if reviews:
            top_comment = reviews[0:4]
            product_details['Top Comment'] = top_comment
        else:
            product_details['Top Comment'] = float('NaN')

    except TimeoutException:
        print("TimeoutException: The top comment element was not loaded in time.")
        product_details['Top Comment'] = float('nan')

    except NoSuchElementException:
        print("NoSuchElementException: The top comment element was not found.")
        product_details['Top Comment'] = float('nan')

    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        product_details['Top Comment'] = float('nan')


    # Extract the image URLs
    try:
        # Wait for the page to load fully
        WebDriverWait(driver, 40).until(
            EC.presence_of_element_located((By.TAG_NAME, 'img'))
        )
        
        # Fetch all image elements on the page
        all_images = driver.find_elements(By.TAG_NAME, 'img')
        
        # Filter images that are likely product images based on their attributes
        product_images = [img.get_attribute('src') for img in all_images if 'alt' in img.get_attribute('outerHTML').lower()]

        selected_images = product_images[:5] if len(product_images) >= 5 else product_images

        # Store the image URLs in the dictionary
        product_details['Image URLs'] = ', '.join(selected_images) if selected_images else 'Images not found'

    except NoSuchElementException:
        product_details['Image URLs'] = 'Images not found'
    except Exception as e:
        product_details['Image URLs'] = f"Error: {str(e)}"

```

```python
# Global variables
page_count = 1
device_count = 0
product_details_list = []

# Define the scraping function
def simulate_product_scraping(max_pages=100):
    global page_count, device_count, product_details_list

    while page_count <= max_pages:
        print(f"Scraping page {page_count}...")

        # Wait for the page to load and locate the product links
        try:
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, "//h2/a[@href]"))
            )
        except TimeoutException:
            print(f"Timeout: Could not locate products on page {page_count}.")
            break

        # Collect all product URLs on the current page
        product_urls = [product.get_attribute('href') for product in driver.find_elements(By.XPATH, "//h2/a[@href]")]
        print(f"Collected {len(product_urls)} product URLs on page {page_count}")

        # Loop through each URL and visit the product page
        for product_url in product_urls:
            try:
                device_count += 1
                print(f"Processing device {device_count} on page {page_count}")

                # Visit the product page
                driver.get(product_url)

                # Initialize a dictionary to store product details
                product_details = {}

                # Run the functions in parallel to scrape content, info, and price
                with concurrent.futures.ThreadPoolExecutor() as executor:
                    futures = [
                        executor.submit(Get_Pro_Price, product_details),
                        executor.submit(Get_Content, product_details),
                        executor.submit(Get_Pro_Info, product_details)
                    ]
                    for future in concurrent.futures.as_completed(futures):
                        future.result()  # Wait for all functions to complete

                # Debugging: Print the brand name
                if 'Brand' in product_details:
                    print(f"Scraped Brand: {product_details['Brand']}")
                else:
                    print(f"Brand not found for device {device_count}")

                # Append the product details to the list
                product_details_list.append(product_details)
                time.sleep(5)  # Add a short pause before the next iteration
                # Go back to the search results page
                driver.back()

            except TimeoutException:
                print("Skipping this product...it's not a mobile phone or took too long.")
                continue
            except Exception as e:
                print(f"An error occurred: {e}")
                continue

        # Attempt to navigate to the next page
        try:
            time.sleep(9)  # Wait for the next page to load
            # Wait until the next button is clickable
            WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[contains(@aria-label, 'Go to next page')]"))
            )
            
            next_button = driver.find_element(By.XPATH, "//a[contains(@aria-label, 'Go to next page')]")
            
            # Check if the next button is not disabled
            if "s-pagination-disabled" not in next_button.get_attribute('class'):
                next_button.click()
                page_count += 1
                print(f"Next button found. Navigating to the next page... Page {page_count}")
                time.sleep(9)  # Wait for the next page to load
            else:
                print("Next button is disabled or no more pages.")
                break

        except NoSuchElementException:
            print("No more pages or pagination not found.")
            logging.info("No more pages or pagination not found.")
            break
        except Exception as e:
            print(f"An error occurred while navigating to the next page: {e}")
            logging.error(f"An error occurred while navigating to the next page: {e}")
            break
```
